```python
import pandas as pd
import numpy as np
```

### create a Pivot table with multiple indexes

```python
df = pd.read_excel('SaleData.xlsx')
df.head()
```

```python
pd.pivot_table(df,index=["Region","SalesMan"])
```

### create a Pivot table and find the total sale amount region wise, manager wise

```python
pd.pivot_table(df,index = ["Region","Manager"], values = ["Sale_amt"],aggfunc=np.sum)
```

### create a Pivot table and find the total sale amount region wise, manager wise, sales man wise

```python
pd.pivot_table(df,index=["Region","Manager","SalesMan"], values="Sale_amt", aggfunc=np.sum)
```

### create a Pivot table and find the item wise unit sold

```python
pd.pivot_table(df,index=["Item"], values="Units", aggfunc=np.sum)
```

### find the region wise total sale.

```python
pd.pivot_table(df,index="Region",values="Sale_amt", aggfunc = np.sum)
```

### region wise, item wise unit sold

```python
pd.pivot_table(df,index=["Region", "Item"], values="Units", aggfunc=np.sum)
```

### create a Pivot table and count the manager wise sale and mean value of sale amount

```python
pd.pivot_table(df,index=["Manager"],values=["Sale_amt"],aggfunc=[np.mean,len])
```

### create a Pivot table and find manager wise, salesman wise total sale and also display the sum of all sale amount at the bottom

```python
table = pd.pivot_table(df,index=["Manager","SalesMan"],values=["Units","Sale_amt"],
               aggfunc=[np.sum],fill_value=0,margins=True)

table
```

### create a Pivot table and find the total sale amount region wise, manager wise, sales man wise where Manager = "Douglas"

```python
table = pd.pivot_table(df,index=["Region","Manager","SalesMan"], values="Sale_amt")
table
```

```python
table.query('Manager == ["Douglas"]')
```

### create a Pivot table and find the region wise Television and Home Theater sold

```python
table = pd.pivot_table(df,index=["Region", "Item"], values="Units")
table
```

```python
table.query('Item == ["Television","Home Theater"]')
```

### create a Pivot table and find the minimum sale value of the items

```python
table = pd.pivot_table(df, index='Item', values='Sale_amt', aggfunc=np.min)
table
```
